1. Solve the wave equation `dt^2 u(t,x) = dx^2 u(t,x)` numerically. Use a `1+1`-dimensional domain, and use Dirichlet boundary conditions `u(t,-L) = u(t,+L) = 0`. Use the spatial domain bounds `[-1; +1]`, and use the Gaussian-like function `f(t,x) = A exp(-1/2 sin(2pi (x-t))^2 / W^2)` as initial condition. Use `L=1, A=1, W=0.2` as parameters. Evolve from `t=0` to `t=4L`, i.e. for two crossing times. Plot the result.

```julia
# evolve_step takes in u of size (xN, 2) and returns a vector of size(xN) as the next time step.
# dt^2 u = (-2 * u(t) + u(t-dt) + u(t+dt)) / dt^2
#        = dx^2 u
#        = (-2 * u(x) + u(x+dx) + u(x-dx)) / dx^2
#
# u(t+dt) = 2*u(t) - u(t-dt) + (- 2*u(x) + u(x+dx) + u(x-dx)) * dt^2 / dx^2
#
function evolve_step(u, dt, dx)
    xN = size(u)[1]
    new_u = zeros(xN)
    
    for i=2:xN-1
        new_u[i] = 2 * u[i, 2] - u[i, 1] + (-2 * u[i,2] + u[i+1,2] + u[i-1,2]) * dt^2 / dx^2
    end
    
    return new_u    
end


function evolve(u_init, dt, dx, t_end)
    ts = -dt:dt:t_end
    us = zeros(size(u_init)[1], length(ts))
    
    us[:, 1:2] = u_init
    for i=3:size(us)[2]
        us[:,i] = evolve_step(us[:,i-2:i-1], dt, dx)
    end
    
    return us
end
```

```julia
L = 1
A = 1
W = 0.2
t_end = 4

xN = 1000
courant = 0.5

xs = range(-L, L, xN)
dx = xs.step.hi

dt = courant * dx

f(t,x) = A *exp(-1/2 * sin(2*pi* (x-t - 0.25))^2 / W^2) # f is altered slightly so as to avoid putting the Gaussian bumps right on the boundaries
u_init = [f(t, x) for x in xs, t in (-dt, 0)]

us = evolve(u_init, dt, dx, t_end)
```

```julia
heatmap(xs, -dt:dt:t_end, us')
```

2. Vary the time step size `dt`. If you choose `dt` too large, the discrete system will be unstable. What is the stability limit? Why does this happen? How could you avoid this limit?

```julia
L = 1
A = 1
W = 0.2
t_end = 4

xN = 1000
courant = 1.00002

xs = range(-L, L, xN)
dx = xs.step.hi

dt = courant * dx

f(t,x) = A *exp(-1/2 * sin(2*pi* (x-t - 0.25))^2 / W^2) # f is altered slightly so as to avoid putting the Gaussian bumps right on the boundaries
u_init = [f(t, x) for x in xs, t in (-dt, 0)]

us = evolve(u_init, dt, dx, t_end)
heatmap(xs, -dt:dt:t_end, us')
```

The Courant limit is at `courant = 1`.  For this particular setup, the system still looks pretty good at `courant = 1`, but begins to break down at `courant = 1.00002`. This could be surmounted by using a higher-order method, or an implicit method.

3. The energy density of a scalar wave is `epsilon(t,x) = 1/2 ((d/dt u(t,x))^2 + ((d/dx u(t,x))^2)`. Calculate the energy density as a function of space and time for your solution. Plot the result.

```julia
function epsilon(us, dt, dx)
    energy = zeros(size(us) .- 1)
    
    udx = (us[1:end-1,:] .- us[2:end,:]) / dt
    udt = (us[:,1:end-1] .- us[:,2:end]) / dx
    
    # Taking some averages to fix the sizes of these arrays
    udx = udx[:,2:end]#(udx[:,1:end-1] + udx[:,2:end])/2
    udt = (udt[1:end-1,:] + udt[2:end,:])/2
    
    return (udx .^ 2 .+ udt .^ 2)/2
end
```

```julia
xN = 1000

xs = range(-L, L, xN)
dx = xs.step.hi

courant = 0.9

dt = courant * dx
ts = -dt:dt:t_end

f(t,x) = A *exp(-1/2 * sin(2*pi* (x-t - 0.25))^2 / W^2) # f is altered slightly so as to avoid putting the Gaussian bumps right on the boundaries
u_init = [f(t, x) for x in xs, t in (-dt, 0)]

us = evolve(u_init, dt, dx, t_end)
#heatmap(xs, -dt:dt:t_end, us')
heatmap(xs[1:end-1] .+ dx/2, ts[1:end-1] .+ dt/2, epsilon(us', dt, dx))
```

4. To calculate the total energy `E(t)` as a function of time, you need to integrate it over space. One way to do so is using the function below. Calculate and show the total energy `E(t)` as a function of time. How well is the energy conserved? How does this depend on the grid spacings `dx` and `dt`?

```julia
# Integrate `epsilon` over space
function integrate(eps, n, dx)
    #println(size(eps))
    ni, nt = size(eps)
    # The boundary points are special; they have a weight of 1/2.
    # There are `ni` points overall, and the grid spacing is `dx = 2L/(ni-1)`.
    # Integrating the constant function `epsilon=1` thus gives exactly `2L`.
    E = eps[1, n] / 2
    for i in 2:ni-1
        E += eps[i, n]
    end
    E += eps[ni, n] / 2
    # Scale by the grid spacing
    E *= dx
    return E
end
```

```julia
ueps = epsilon(us', dt, dx)
uenergies = [integrate(ueps', i, dx) * (dt / dx)^2 for i in 1:length(ts)-2]

plot(ts[2:end-1], uenergies)
```

This is a very narrow range!  And it only varies when two wavepackets are colliding with one another.  The fact that it always comes back to the same number in between collisions makes me confident that things are going well here.
